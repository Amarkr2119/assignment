# CORRELATION MATRIX

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("Book_ipl22_ver_33.csv")


corr_matrix = df.corr(numeric_only=True)


plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

# SCATTER PLOT WITH CATEGORIES (First Innings Score & second Innings Score and Categorical Hue: "venue" )

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("Book_ipl22_ver_33.csv")


plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["first_ings_score"], y=df["second_ings_score"], hue=df["venue"], palette="viridis")

plt.title("Scatter Plot: First Innings Score vs. Second Innings Score by Venue")
plt.xlabel("First Innings Score")
plt.ylabel("Second Innings Score")
plt.legend(title="Venue", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

#STACKED COLUMN CHART(VENUE AND TOSS DECISION

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Book_ipl22_ver_33.csv")


stacked_data = df.pivot_table(index="venue", columns="toss_decision", aggfunc="size", fill_value=0)

stacked_data.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")

plt.title("Stacked Column Chart: Toss Decision by Venue")
plt.xlabel("Venue")
plt.ylabel("Match Count")
plt.xticks(rotation=90)
plt.legend(title="Toss Decision")
plt.show()
```
